```python
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
#import pandas as pd
#import xml.etree.ElementTree as ET

import aerios as ae
```

```python
history_file = open('../history_A7-BCL.dat', 'r')
values = []
for line in history_file:
    values.append(line.split(','))
data = values[3]
print(data)
```

```python
origin = ae.Airport(data[2])
destination = ae.Airport(data[3])
```

```python
plane = ae.Airline(airplane_registration=data[4], airports=[origin, destination])
print(origin.country, origin.airport_name, origin.city)
print(plane.airline, plane.aircraft_type, plane.status, plane.engine_id, \
      plane.number_of_engines, plane.first_flight)
```

```python
coast_color = (20.0/255.0, 20.0/255.0, 20/255.0, 0.3)
bg_color = (0.0, 0.0, 0.0, 0.0)

plt.figure(figsize=(14, 10))
m = Basemap(projection='mill',llcrnrlat=-20,urcrnrlat=50,\
            llcrnrlon=0,urcrnrlon=90,resolution='c')

m.drawcoastlines(color=coast_color, linewidth=1.0)
#m.fillcontinents(color=bg_color, lake_color=bg_color)
m.drawmapboundary(fill_color=bg_color)
#m.fillcontinents()
m.shadedrelief()
m.drawparallels(np.arange(10,90,20),labels=[1,1,0,1])
m.drawmeridians(np.arange(-180,180,30),labels=[1,1,0,1])
x1, y1 = m(origin.longitude, origin.lattitude)
x2, y2 = m(destination.longitude, destination.lattitude)


plt.plot(x1, y1, 'ok', markersize=5)
plt.text(x1, y1,  origin.airport_name, fontsize=12)

plt.plot(x2, y2, 'ok', markersize=5)
plt.text(x2, y2, destination.airport_name, fontsize=12)


line, = m.drawgreatcircle(origin.longitude, origin.lattitude, \
                        destination.longitude, destination.lattitude, \
                                  linewidth=3.0, color='crimson')

plt.show()
```
